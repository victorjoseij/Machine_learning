```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

Each row corresponds to the item bought by one customer in one invoice

```python
df = pd.read_csv(r"/content/archive (2).zip")
df
```

```python
items = df["shrimp"].unique()
items
```

column names are just items so renanme columns cuz why not!

```python
new_columns = [f'Column {i}' for i in range(1, len(df.columns) + 1)]
df.columns = new_columns
df
```

generating list odf transactions

```python
transactions=[]
for index,row in df.iterrows():
  transactions.append(row.dropna().tolist())
transactions
```

transform the transactions list into binary matrix format and store it in a dataframe

```python
te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
transaction_table = pd.DataFrame(te_ary, columns=te.columns_)
transaction_table.head()
```

use apriori algorithm to generate all the frequent occuring itemsets

```python
frequent_itemsets = apriori(transaction_table, min_support=0.01, use_colnames=True)
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))
frequent_itemsets
```

checking for the most occuring popular items

```python
popular_items = frequent_itemsets[frequent_itemsets['length'] == 1]
popular_items.drop("length", axis = 1).sort_values(by = 'support', ascending = False).head(10)
```

top 10 most bought items
bar garph visualization

```python
plt.figure(figsize=(10, 15))
labels = popular_items['itemsets'].apply(lambda itemset: ', '.join(itemset))
support_values = popular_items['support'].values
plt.barh(labels, support_values)
plt.xlabel('Popularity')
plt.ylabel('Items')
plt.show()
```

pie chart

```python
total_support = sum(support_values)
percentages = [(support / total_support) * 100 for support in support_values]
plt.figure(figsize=(20, 20))
plt.pie(percentages, startangle=140)
plt.legend(labels, loc='upper left')
plt.show()
```

generate association rules

```python
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.3)
rules
```

```python
rules = rules.drop(labels = ["antecedent support", "consequent support", "leverage", "zhangs_metric"], axis = 1)
```

relationship between different measures

```python
columns_to_plot =['support', 'confidence', 'lift', 'conviction']
correlation_matrix = rules[columns_to_plot].corr()
plt.figure(figsize=(8,7))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", center=0)
plt.show()
```

```python
rules.sort_values(by="support", axis = 0, ascending=False).head(10)
```

the confidence of a rule A->C is the probability of seeing the consequent in a transaction given that it also contains bthe antecedent

```python
rules.sort_values(by="confidence", axis = 0, ascending=False).head(10)
```

The life metric is commonly used to measure how much more often the antecedent and consequent of a rule A->C occur together than we would expect if they were statistically independent,the lift score will be exactly 1

```python
rules.sort_values(by="lift", axis = 0, ascending=False).head(10)
```

A high conviction value means that the conmsequent is highly depending on the antecedent

```python
rules.sort_values(by="conviction", axis = 0, ascending=False).head(10)
```

```python
rules[ (rules['confidence'] > 0.4) &
       (rules['lift'] > 1.8) &
       (rules['conviction'] > 1.2)]
```
